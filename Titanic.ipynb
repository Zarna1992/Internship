```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#dropping Cabin as more than 50% Null values are available
#replacing nul value of age with mean
```

```python
df=df.drop(columns='Cabin',axis=1)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
categorical=[]
for i in df.dtypes.index:
    if df.dtypes[i]=="object":
        categorical.append(i)
print("Categorical columns:",categorical)
print("\n")

numerical=[]
for i in df.dtypes.index:
    if df.dtypes[i]!="object":
        numerical.append(i)
print("Numerical columns:",numerical)
print("\n")
```

```python
#Now we can replace Survived and Embarked to numerical column
```

```python
df.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
df.head()
```

```python
dfnew=df.drop(['PassengerId','Name','Ticket'], axis=1)
dfnew.head()
```

```python
plt.figure(figsize=(15,10))  
sns.heatmap(dfnew.corr())
sns.heatmap(dfnew.corr(), annot = True, vmin=-1, vmax=1, center= 0, cmap= 'Blues', linewidths=3, linecolor='black')

plt.show()
```

```python
# Spliting of data into training(80%) and testing(20%) sets

from sklearn.model_selection import train_test_split
```

```python
#taking high correlation value
# X = features, y = target variable

X = dfnew[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = dfnew['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler 
scale=StandardScaler()
dfnew=scale.fit_transform(dfnew)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
actual_vs_pred=pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
actual_vs_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_test_prediction=lr.predict(X_test)
test_data_accuracy=accuracy_score(y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
